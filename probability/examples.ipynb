```python
import matplotlib.pyplot as plt
import random as r

N = 1000
xs = [x for x in range(N)]
ys = [r.random() for x in xs]
fig, ax = plt.subplots()
ax.scatter(xs, ys, marker='.', label="random")
plt.axhline(0.5, color="green", label="0.5")
plt.ylim(0, 1)
plt.legend()
```

```python
xs = [x for x in range(N)]
tmp = [r.random() for x in xs]
ys = [y0*y1 for (y0,y1) in zip(tmp, tmp)]
fig, ax = plt.subplots()
ax.scatter(xs, ys, marker='.', label="random * random")
plt.axhline(0.5, color="green", label="0.5")
plt.ylim(0, 1)
plt.legend()
```

```python
xs = [x for x in range(N)]
tmp = [r.random() for x in xs]
ys = [(y0+y1)/2.0 for (y0,y1) in zip(tmp, tmp)]
fig, ax = plt.subplots()
ax.scatter(xs, ys, marker='.', label="random * random")
plt.axhline(0.5, color="green", label="0.5")
plt.ylim(0, 1)
plt.legend()
```
